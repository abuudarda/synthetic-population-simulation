```python
import pandas as pd
import numpy as np
import random
import itertools

col_list = ["SERIALNO","SEX", "RAC1P", "AGEP","SCHL","HINCP","VEH","NP","WIF"]
df = pd.read_csv("C:\\Users\\abubabu\\Documents\\GitHub\\synthetic-population-simulation\\CA-data\\pums\\temp_pums.csv", usecols=col_list)

df.head(5)
```

```python
df.isnull().sum()
```

```python
df['VEH']=df['VEH'].fillna(df['VEH'].median())
df['NP']=df['NP'].fillna(df['NP'].median())
df['WIF']=df['WIF'].fillna(df['WIF'].median())
```

```python
list(df.columns)
```

```python
df.head(5)
```

```python
from itertools import product
household_columns = ["HINCP","VEH","NP","WIF"]
person_columns = ["AGEP","SCHL","SEX","RAC1P"]
household_combinations = []
person_combinations = []
hh_unique_values = {col: df[col].unique() for col in household_columns}
p_unique_values = {col: df[col].unique() for col in person_columns}
for combo in product(*hh_unique_values.values()):
    formatted_combo = '_'.join(f"{value}{col}" for col, value in zip(household_columns, combo))
    household_combinations.append(formatted_combo)
for combo in product(*p_unique_values.values()):
    formatted_combo = '_'.join(f"{value}{col}" for col, value in zip(person_columns, combo))
    person_combinations.append(formatted_combo)
household_combinations.sort()
person_combinations.sort()
len(household_combinations),len(person_combinations)
```

```python



# Get unique households
unique_households = df['SERIALNO'].unique()

# Initialize the binary matrix
num_unique_households = len(unique_households)
num_household_combinations = len(household_combinations)
num_person_combinations = len(person_combinations)
binary_matrix = pd.DataFrame(0, index=unique_households, columns=household_combinations + person_combinations)

# Fill the binary matrix
for idx, row in df.iterrows():
    serialno = row['SERIALNO']
    household_combination = "{}HINCP_{}VEH_{}NP_{}WIF".format(row["HINCP"], row["VEH"], row["NP"], row["WIF"])
    person_combination = "{}AGEP_{}SCHL_{}SEX_{}RAC1P".format(row["AGEP"], row["SCHL"], row["SEX"], row["RAC1P"])

    binary_matrix.at[serialno, household_combination] = 1
    binary_matrix.at[serialno, person_combination] += 1  # Increase the frequency

# Save the binary matrix to a CSV file
binary_matrix.to_csv('binary_matrix.csv')
```
