```python
residential_data_reset = residential_areas.reset_index(drop=True)
residential_counter = 0
for idx, row in sampled_data.iterrows():
    residential_row = residential_data_reset.iloc[residential_counter]
    
    # Copy 'LAT' and 'LON' values from residential_row to the sampled_data row
    sampled_data.at[idx, 'LAT'] = residential_row['LAT']
    sampled_data.at[idx, 'LON'] = residential_row['LON']
    
    # Increment the counter for the next iteration, and reset to 0 if it exceeds the length of residential_data
    residential_counter = (residential_counter + 1) % len(residential_data_reset)
```

```python
sampled_data=sampled_data.loc[sampled_data.index.repeat(sampled_data['HHSIZE'])]
sampled_data.reset_index(drop=True, inplace=True)
# sampled_data.head(39)
```

```python
data_to_predict=sampled_data
data_to_predict.head()
```

```python
def translateEduc(educ):
    if educ == 'Graduate':
        return 5
    elif educ == 'Bachelor':
        return 4
    elif educ == 'college':
        return 3  # Assigning a common code for both college and some college/vocational
    elif educ == 'High-school':
        return 2
    else:
        return 1
    
def translateAge(age):
    if age =='child':
        return 1
    elif age == 'tenager':
        return 2
    elif age == 'young-adult':
        return 3
    elif age == 'adult':
        return 4
    else:
        return 5
def translateSex(sex):
    if sex =='male':
        return 1
    else:
        return 2
```

```python
data_to_predict['EDUC'] = data_to_predict.apply(lambda row: translateEduc(row['EDUC']), axis=1)
data_to_predict['R_AGE'] = data_to_predict.apply(lambda row: translateAge(row['R_AGE']), axis=1)
data_to_predict['R_SEX'] = data_to_predict.apply(lambda row: translateSex(row['R_SEX']), axis=1)
# data_to_predict['WRKCOUNT'] = data_to_predict.apply(lambda row: translatewc(row['WRKCOUNT']), axis=1)
data_to_predict['WRKCOUNT']=data_to_predict['WRKCOUNT'].fillna(0)
```
