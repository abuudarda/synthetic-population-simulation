```python
import pandas as pd;

# construct age categories from numerical ages
def translateAge(age):
    if age <=5:
        return 'child'
    elif age <= 18:
        return 'tenager'
    elif age <= 25:
        return 'young-adult'
    elif age < 60:
        return 'adult'
    else:
        return 'senior'

# get the symbolic name of the gender from numerical value
def translateSex(sex):
    if sex == 1:
        return 'male'
    else:
        return 'female'

# get the race categories from numerical race values
def translateRace(race):
    if race == 1:
        return 'white'
    elif race == 2:
        return 'black'
    elif race >= 3 and race <= 5:
        return 'native'
    elif race == 6:
        return 'asian'
    elif race == 7:
        return 'pacific-islander'
    else:
        return 'other'
    

def translateEduc(educ):
    if educ>=22 and educ<=24: return 'Graduate'
    elif educ==21: return 'Bachelor'
    elif educ>=18 and educ<=20: return 'college'
    elif educ>=16 and educ<=17: return 'High-school'
    else : return 'below-high'

def translateInc(inc):
    if inc>1 and inc<1000: return "1"
    elif inc<15000 : return '2'
    elif inc<25000 : return '3'
    elif inc<35000 : return '4'
    elif inc<50000 : return '5'
    elif inc<75000 : return '6'
    elif inc<100000 : return '7'
    # elif inc<125000 : return '8'
    elif inc<150000 : return '8'
    elif inc<200000 : return '9'
    elif inc>=200000 : return '10'
    else : return '0'
```

```python
col_list = ["SEX", "RAC1P", "AGEP","SCHL", "SERIALNO"]
pums_data = pd.read_csv("psam_p46.csv",usecols=col_list)
hh_data = pd.read_csv("psam_h46.csv", usecols=["SERIALNO","HINCP","VEH","NP","WIF"])
pums_data=pums_data.merge(hh_data[["SERIALNO","HINCP","VEH","NP","WIF"]],
                          on="SERIALNO", how='left')
pums_data['VEH'].value_counts()
```

```python
# only read the columns that we are interested as the CSV file is too large
# col_list = ["SEX", "RAC1P", "AGEP","SCHL", "SERIALNO"]
# pums_data = pd.read_csv("psam_p46.csv", usecols=col_list)
# hh_data = pd.read_csv("psam_h46.csv", usecols=["SERIALNO"])

# construct caegorical data from numerical data that comes from pums
pums_data['AGEP'] = pums_data.apply(lambda row: translateAge(row['AGEP']), axis=1)
pums_data['SEX'] = pums_data.apply(lambda row: translateSex(row['SEX']), axis=1)
pums_data['RAC1P'] = pums_data.apply(lambda row: translateRace(row['RAC1P']), axis=1)
pums_data['SCHL'] = pums_data.apply(lambda row: translateEduc(row['SCHL']), axis=1)
pums_data['HINCP'] = pums_data.apply(lambda row: translateInc(row['HINCP']), axis=1)
# pums_data['house_no'] = pums_data.apply(lambda row: translateRace(row['SERIALNO']), axis=1)


# view the data to see if the categorization is accurate
pums_data.head(20)
# hh_data.columns.tolist()
```
